```python
import numpy as np
import matplotlib.pyplot as plt

methods = [
    "Wu et al. (2022)\nRF  NSL-KDD",
    "Chen et al. (2021)\nADASYN+RF  CICIDS-2017",
    "This work\nRF+SMOTE  HIKARI-2021"
]

metrics = ["Accuracy", "Precision", "Recall", "F1-score", "AUC"]

values = np.array([
    [0.9834, 0.9838, 0.9815, 0.9826,    np.nan],   # Wu et al. (2022)
    [np.nan, 0.9850, 0.9230, 0.9530,    0.9978],   # Chen et al. (2021)
    [0.8538, 0.6654, 0.9207, 0.7058,    0.9420]    # This work
])

values_pct = values * 100.0

x = np.arange(len(metrics))
width = 0.25

fig, ax = plt.subplots(figsize=(9, 5))

for i, method in enumerate(methods):
    ax.bar(x + (i - 1) * width, values_pct[i], width, label=method)

ax.set_ylabel("Score (%)")
ax.set_xticks(x)
ax.set_xticklabels(metrics)

# *** Start y-axis at 0 instead of 60 ***
ax.set_ylim(0, 100.5)

ax.set_title("Performance Comparison of Random ForestBased IDS Approaches")

# Move legend to the side so it doesnt cover bars
ax.legend(loc="center left", bbox_to_anchor=(1.02, 0.5), borderaxespad=0.)

# Add value labels
for i in range(values_pct.shape[0]):
    for j in range(values_pct.shape[1]):
        val = values_pct[i, j]
        if np.isnan(val):
            continue
        ax.text(
            x[j] + (i - 1) * width,
            val + 1,
            f"{val:.1f}",
            ha="center",
            va="bottom",
            fontsize=8
        )

fig.tight_layout()
plt.savefig("rf_ids_comparison.png", dpi=300, bbox_inches="tight")
plt.show()
```
